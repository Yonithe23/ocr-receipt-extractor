```python
# ==============================
# STEP 1: Import Libraries
# ==============================

import easyocr
import json
from datetime import datetime
from pathlib import Path
```

```python
IMAGE_PATH = "../Data/receipt.png"

reader = easyocr.Reader(["en"])

results = reader.readtext(IMAGE_PATH)

ocr_data = {
    "file_name": Path(IMAGE_PATH).name,
    "ocr_engine": "easyocr",
    "processed_at": datetime.now().isoformat(),
    "full_text": "",
    "lines": []
}

full_text_list = []

for bbox, text, confidence in results:
    
    # Convert bbox int32 to normal int
    clean_bbox = [
        [int(point[0]), int(point[1])]
        for point in bbox
    ]

    full_text_list.append(text)

    ocr_data["lines"].append({
        "text": text,
        "confidence": round(float(confidence), 4),
        "bbox": clean_bbox
    })

ocr_data["full_text"] = "\n".join(full_text_list)

print(json.dumps(ocr_data, indent=4, ensure_ascii=False))

with open("ocr_output.json", "w", encoding="utf-8") as f:
    json.dump(ocr_data, f, indent=4, ensure_ascii=False)

print("JSON file saved successfully")
```
