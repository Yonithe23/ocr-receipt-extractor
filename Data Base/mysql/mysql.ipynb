```python
import json
import mysql.connector
```

```python
sample_data = {
    "receipt_type": "VAT RECEIPT",
    "transaction_id": "000456789",
    "date": "11/03/2024",
    "total": "403.65"
}

with open("ocr_output.json", "w", encoding="utf-8") as f:
    json.dump(sample_data, f, indent=4)

print("JSON file created!")
```

```python
with open("ocr_output.json", "r", encoding="utf-8") as f:
    data = json.load(f)

print(data)
```

```python
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="root",
    database="receipt_db"
)

cursor = conn.cursor()

print("Connected to MySQL!")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS receipts (
    id INT AUTO_INCREMENT PRIMARY KEY,
    receipt_type VARCHAR(255),
    transaction_id VARCHAR(255),
    receipt_date VARCHAR(255),
    total VARCHAR(255)
)
""")

conn.commit()

print("Table created!")
```

```python
cursor.execute("""
INSERT INTO receipts (
    receipt_type,
    transaction_id,
    receipt_date,
    total
)
VALUES (%s, %s, %s, %s)
""", (
    data["receipt_type"],
    data["transaction_id"],
    data["date"],
    data["total"]
))

conn.commit()

print("Data inserted!")
```

```python
cursor.execute("SELECT * FROM receipts")

rows = cursor.fetchall()

for row in rows:
    print(row)
```
