```python
from paddleocr import PaddleOCR

ocr = PaddleOCR(use_angle_cls=True, lang='en')

result = ocr.ocr("../Data/receipt.png")

# Extract clean text
full_text = []

for line in result[0]:
    text = line[1][0].strip()   # remove extra spaces
    full_text.append(text)
print("---------------------------")

# Display line by line
for text in full_text:
    print(text)
```

```python
from paddleocr import PaddleOCR

ocr = PaddleOCR(use_angle_cls=True, lang='en')

result = ocr.ocr("../Data/receipt.png")

# Extract text
full_text = []

for line in result[0]:
    text = line[1][0].strip()
    full_text.append(text)

# Convert to Dictionary
receipt_data = {
    "receipt_type": full_text[0],
    "transaction_id": full_text[1],
    "date": full_text[2],
    "subtotal": full_text[11],
    "vat": full_text[15],
    "total": full_text[16]
}

print(receipt_data)

print("-------------------------")
print(receipt_data)
```
